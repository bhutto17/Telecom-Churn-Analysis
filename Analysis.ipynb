# **Customer Churn Analysis, Visualisations & Predictive Modelling**

For this project, the main objective is to assist the company in improving customer retention by identifying high value customers and churn risks.


## **Business Problem**

The data  driven project aims to derive actionable insights to enhance customer retention by focusing on high-value customers and identifying churn risks. The primary concern addressed is the alarming customer churn rate witnessed by the company, leading to significant revenue loss. A higher churn rate suggests a critical problem that demands an in-depth analysis to unravel the underlying causes and patterns. To better understand the customer churn in the Telecom Company, using a detailed analytical approach,

First and foremost, a comprehensive examination of the dataset will be performed, shedding light on facts like customer demographics, their service tenure, billing tendencies, prevalent subscriptions, offers preffered by users, and any additional services they demand. Along with this, data visualization will be of key importance, using a number of different graphs and charts, we’ll be better able to understand the underlying reasons for higher churn rates. These visual will help meticulously in spotlighting trends within dataset for different type of customers and it will make stakeholders to better understand the customers.

Lastly, to make sure our findings are based on solid evidence, we will use statistical tools, specifically using t-tests and regression analyses to test our ideas and better understand the reasons behind customer churn.
